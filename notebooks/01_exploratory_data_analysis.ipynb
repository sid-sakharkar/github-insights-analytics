```python
# Exploratory Data Analysis on GitHub Repository Data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import json

# Function to fetch GitHub repository data
def fetch_github_repo_data(owner, repo):
    url = f'https://api.github.com/repos/{owner}/{repo}'
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        return None

# Fetch data for a specific repository
owner = 'octocat'  # Example owner
repo = 'Hello-World'  # Example repo
repo_data = fetch_github_repo_data(owner, repo)

if repo_data:
    # Convert to DataFrame
    df = pd.json_normalize(repo_data)
    print(df.head())

    # Visualize data
    plt.figure(figsize=(12,6))
    sns.barplot(x='stargazers_count', y='name', data=df)
    plt.title('Repository Stars')
    plt.xlabel('Number of Stars')
    plt.ylabel('Repository Name')
    plt.show()
else:
    print('Failed to fetch repository data')
```
